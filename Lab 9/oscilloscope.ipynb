# Lab 9: Oscilloscope

$\textbf{Objective}$

The objective of this experiment is to measure the frequency of light intensity variations from three different light sources: an incandescent bulb, a Compact Fluorescent Light (CFL), and a Light Emitting Diode (LED). We will use a photodiode and an oscilloscope to analyze the waveforms and determine which source provides the steadiest light.

$\textbf{Theory}$

Unlike previous experiments involving mechanical motion, this lab focuses on alternating current (AC) and optics. We analyze how different lighting technologies convert AC electricity into light and how this affects the stability of the light output.

**1. AC Power and Frequency Doubling**
Household electricity is Alternating Current (AC) operating at a frequency of 60 Hz. The voltage oscillates between positive and negative peaks. However, the power (and thus brightness) of a light bulb is proportional to the square of the voltage:
$$
P \propto V^2
$$
Since squaring a sine wave turns the negative half-cycles into positive ones, the power peaks twice per voltage cycle. Therefore, we expect the light intensity to oscillate at **120 Hz** (double the mains frequency).

**2. Light Source Characteristics**
* **Incandescent:** Uses a tungsten filament. Due to **thermal inertia**, the filament stays hot even when the voltage drops to zero, preventing the light output from completely shutting off (DC offset).
* **LED:** Uses a diode (semiconductor). Since standard AC switches direction, LEDs require a **rectifier** to convert AC to DC. A capacitor is often added to smooth the output, preventing the light from dropping to zero.
* **CFL:** Uses a **ballast** to regulate current, which often operates at much higher frequencies to prevent visible flicker and improve efficiency.

$\textbf{Experiment Method/Procedure}$

A photodiode was placed near each light source to convert the light intensity into an electrical voltage signal. An oscilloscope was used to capture this signal over time.

The oscilloscope settings were adjusted to capture a clear waveform for each bulb. The data was exported to .csv files for three sources:
1.  Incandescent Bulb
2.  Light Emitting Diode (LED)
3.  Compact Fluorescent Light (CFL)

The data was analyzed to generate Amplitude vs. Time plots. The frequency was determined by measuring the period ($T$) between peaks in the waveform using the relationship $f = 1/T$.

$\textbf{Discussion}$

**Frequency Analysis:**
Both the Incandescent and LED bulbs oscillated at 120 Hz. This confirms the theoretical expectation that light intensity oscillates at double the standard 60 Hz mains frequency because power is proportional to voltage squared ($P \propto V^2$).

**Waveform Characteristics:**
* **Incandescent:** The signal never approached 0 V due to the **thermal inertia** of the tungsten filament. The filament retains significant heat and continues to emit light even when the AC current momentarily drops to zero.
* **LED:** The waveform also did not drop to zero, indicating the presence of a **capacitor** in the circuit. The capacitor stores energy during peak voltage and releases it during the dips, "filling in" the gaps to minimize flicker.
* **CFL:** The CFL operated at much higher frequencies (~343â€“487 Hz). This is due to the electronic ballast, which increases the frequency to stabilize the current.

**Stability:**
The **CFL** provides the steadiest light. Human eyes can perceive flicker up to about 60 Hz. The 120 Hz flicker of the LED and Incandescent is close to this limit and can be perceived by some. The CFL's frequency is far higher, rendering the variations invisible to the human eye.

$\textbf{Conclusion}$

The experiment confirmed that standard AC light sources (Incandescent and LED) oscillate at 120 Hz. The incandescent bulb's flicker is smoothed by thermal inertia, while the LED relies on capacitors. The CFL operates at a much higher frequency, providing the steadiest light source for human perception.
